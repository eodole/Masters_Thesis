```python
#read csv 
pest_df = pd.read_csv("./data/pesticide_data_2010.csv")

# get rid of first row 
pest_df.drop(["Unnamed: 0"],axis=1, inplace=True )




```

```python
np.set_printoptions(threshold=np.inf)
print(np.unique(pest_df["FIPS"]))
print(np.unique(county_df["fips"]))
```

```python
# which ones are missing 
missing_counties = []
FIPS = np.unique(pest_df["FIPS"]) 
for x in np.unique(county_df["fips"]):
    if x not in FIPS: 
        missing_counties.append(x)
```

```python
missing_counties 
```

```python
full_df = pd.merge(county_df, pest_df, how="left", left_on = "fips", right_on = "FIPS")
```
